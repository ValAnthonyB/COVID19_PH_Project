```python
import numpy as np
import pandas as pd
```

```python
# Timestamp

from datetime import date

today = date.today() # date object

# Formatted time stamp
time_stamp = today.strftime("%m-%d-%Y")
```

```python
url='https://coronavirus-ph-api.herokuapp.com/cases' # From API

df = pd.read_json(url) # Getting the data
df.set_index('case_no', inplace=True)

#df.drop(['latitude', 'longitude', 'resident_of'], axis = 1, inplace=True)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    display(df)
```

```python
# Renaming the date into date_of_death_recovery

new_cols = df.columns.values
new_cols[0] = 'date_of_death_recovery'
df.columns = new_cols
```

```python
df.head()
```

# Correcting the dates

```python
# From wikipedia March 29
url_wiki = 'https://web.archive.org/web/20200330075023/https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_the_Philippines'

df_wiki = pd.read_html(url_wiki)[4]
df_wiki.set_index('Caseno.', inplace=True)
df_wiki = df_wiki.loc[~df_wiki.index.duplicated(keep='first')]
df_wiki.head()

date_announce = list(df_wiki.Date.copy())

# Corrections
march30 = ['March 30' for i in range(0,128)]
march31 = ['March 31' for i in range(0,538)]

date_announce = date_announce + march30 + march31
```

```python
df['date_announce'] = date_announce
```

```python
# Changing order of rows

cols = ['date_announce',
             'age',
             'gender',
             'nationality',
             'hospital_admitted_to',
             'travel_history',
             'status', 
             'date_of_death_recovery']

df = df[cols]
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    display(df)
```

# Cleaning the addresses

```python
import re
```

```python
ls_hospitals = np.unique(df.hospital_admitted_to)
```

```python
for i in range(0,len(ls_hospitals)):
    hospital = ls_hospitals[i]
    if ', Inc' in hospital:
        corrected = re.findall(r'(.+), Inc.', hospital)[0]
        print(corrected)
    
        # Replacement
        df.hospital_admitted_to.replace(to_replace = ls_hospitals[i], 
                     value=corrected, inplace=True) 
        ls_hospitals[i] = corrected
    else:
        pass
```
