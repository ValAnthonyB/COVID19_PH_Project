```python
# Test

g = 'Uni - Health Southwoods Hospital And Medical Center, Inc.'
lst = re.findall(r'(.+), Inc.', g)
lst[0]
```

```python
for i in range(0,len(ls_hospitals)):
    hospital = ls_hospitals[i]
    if ', Inc' in hospital:
        corrected = re.findall(r'(.+), Inc.', hospital)[0]
        print(corrected)
    
        # Replacement
        df.hospital_admitted_to.replace(to_replace = ls_hospitals[i], 
                     value=corrected, inplace=True) 
        ls_hospitals[i] = corrected
    else:
        pass
```

```python
import geopy
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim
```

```python
# Test

geolocator = Nominatim(user_agent="vv")
geolocator.geocode('Lucena United Doctors General Hospital').address
```

```python
geolocator = Nominatim(user_agent="vv")

#recursive function in case a timeout occurs
def do_geocode(address):
    try:
        return geolocator.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)

# location = geolocator.geocode('The Medical City–Pasig')
# print(location.address)
```

```python
addresses = []
latitudes = []
longitudes = []

for i, hospital in enumerate(np.unique(df.hospital_admitted_to)):    
    if hospital!='eyyyyyy': #disregard CDJ Hospital and the TBAs
        location = do_geocode(hospital)
        print(i, location.address)
        
        # Appending the address, lat and longs
        addresses.append(location.address)
        latitudes.append(location.latitude)
        longitudes.append(location.longitude)
```

```python
np.sum(np.array(addresses)=='NA') # if only 0, the errors are only CDJ hospital and For Validation
```

```python
provinces = []

for i, province in enumerate(addresses):
    prov=list(province.split(","))
    if 'Metro Manila' in province:
        print(i, 'Metropolitan Manila')
        provinces.append('Metropolitan Manila')
        
    if 'Metro Manila' not in province:
        # Missing postal codes
        if 'Cavite Medical Center' in province or 'President Ramon Magsaysay Memorial Hospital' in province:
            print(i, prov[-3].lstrip())
            provinces.append(prov[-3].lstrip())
            continue
        
        # Address doesnt seem to contain province
        if 'Southern Philippines Medical Center' in province:
            print('Davao del Sur')
            provinces.append('Davao del Sur')
            continue
        
        # Address doesnt seem to contain province
        if 'West Metro Medical Center' in province:
            print('Zamboanga del Sur')
            provinces.append('Zamboanga del Sur')
            continue
        else:
            print(i, prov[-4].lstrip())
            provinces.append(prov[-4].lstrip())
```
