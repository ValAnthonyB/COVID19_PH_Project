### Geocoding

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

geolocator = Nominatim(user_agent="vv")
```

```python
def do_geocode(g):
    try:
        print(geolocator.geocode(g).address)
        return geolocator.geocode(g)
    except GeocoderTimedOut:
        return do_geocode(g)
```

```python
df['coordinates'] = df['health_facility'].apply(do_geocode).apply(lambda x: (x.latitude, x.longitude))
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    display(df)
```

# Folium

```python
import folium
from folium.plugins import MarkerCluster
```

```python
lats = [df.coordinates[i][0] for i in range(0, len(df.coordinates))]
longs = [df.coordinates[i][1] for i in range(0, len(df.coordinates))]
```

```python
m = folium.Map(location=[np.mean(lats), np.mean(longs)], zoom_start=5.45)
mc = MarkerCluster()

for point in range(0, len(df)):
    folium.Marker(df.coordinates[point], popup=df.health_facility[point]).add_to(mc)

m.add_child(mc)
m
```
